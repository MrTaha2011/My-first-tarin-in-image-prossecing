# Import librarys

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from scipy import ndimage
```

# edge finder

input image :

tarin2.png

```python
roberts_cross_v = np.array( [[1, 0 ],
							[0,-1 ]] )

roberts_cross_h = np.array( [[ 0, 1 ],
							[ -1, 0 ]] )

img = cv2.imread("/content/tarin2.pdf.png",0).astype('float64')
img/=255.0
vertical = ndimage.convolve( img, roberts_cross_v )
horizontal = ndimage.convolve( img, roberts_cross_h )

edged_img = np.sqrt( np.square(horizontal) + np.square(vertical))
edged_img*=255
cv2.imwrite("output.jpg",edged_img)

cv2_imshow(a = edged_img)
```

# Put text and change colors to black and white

input image :

ud.jpg

```python
import cv2

img = cv2.imread('/content/train1.pdf.png')

# cv2.imshow('Taha', img)
# cv2.waitKey(0)

font  = cv2.FONT_HERSHEY_COMPLEX
point = (400, 214)
color = (0, 0, 0)
scale = 0.5
line_type = 2

cv2.putText(img, 'Taha Bayat',point,font,scale,color,line_type)

black_white = cv2.cvtColor(src=img,code=cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
print("")
cv2_imshow(black_white)
# cv2.waitKey(0)

cv2.imwrite('blackw.jpg', black_white)
```
